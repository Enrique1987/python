```
from fuzzywuzzy import process
```

```
data = ['MONTREAL EDUCATION BOARD', 'Ile de Montreal', 'Montreal',
       'Ville de Montreal','banana', 'StLouis', 'St Louis', 'Saint Louis']

treshold = 75
```

```
pepe = process.extract("StLouis", data)
similar_words = []
for j, y in pepe:
    if y > treshold:
        similar_words.append(j)
        print(j, y)
```

```
print(similar_words)
```
